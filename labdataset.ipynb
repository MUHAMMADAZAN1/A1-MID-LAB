```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Choose your features (now using the correctly assigned column names)
X = df[['sepal_length']]   # independent variable (must be 2D)
y = df['petal_length']     # dependent variable

# Fit the model
reg = LinearRegression()
reg.fit(X, y)

m = reg.coef_[0]       # slope
c = reg.intercept_     # intercept

# Plot data + regression line
plt.figure(figsize=(8,6))
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, reg.predict(X), color='red', linewidth=2, label='Regression line')

# Build the equation text
equation = f"y = {m:.2f}x + {c:.2f}"

# Display equation on the graph
plt.text(0.05, 0.95, equation, transform=plt.gca().transAxes,
         fontsize=14, color='black', verticalalignment='top',
         bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Linear Regression: y = mx + c')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("Equation:", equation)
print("Slope (m):", m)
print("Intercept (c):", c)
```
